The first step is generate random data points for prediction. Run the code provided and answer the five questions included in this notebook.

```python
# Do not modify this cell
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

# Question 1
**Q1**: You need to output the shapes of X and y in separate lines using numpy.

- Write your code in the cell below
- Do not delete the comment line ### Your code here ###

```python
### Q1: Your code here ###
print(np.shape(X), "\n")
print(np.shape(y), "\n")
```

We can use matplotlib to visualise the dataset:

```python
# Do not modify this cell
import matplotlib.pyplot as plt
plt.plot(X, y, "b.")
plt.show()
```

We can scale the data using pandas.

```python
# # Do not modify this cell
import pandas as pd

data = {'n_comp_affected': X.flatten(), 'cost': y.flatten()}
df = pd.DataFrame(data)
df['n_comp_affected'] = df['n_comp_affected'] * 1000
df['n_comp_affected'] = df['n_comp_affected'].astype('int')
df['cost'] = df['cost'] * 1000
df['cost'] = df['cost'].astype('int')

# TO ASK: whats the purose of x1000

# Create a scatter plot
plt.scatter(df['n_comp_affected'], df['cost'])

# Label the axes
plt.xlabel('Number of Components Affected')
plt.ylabel('Cost')

# Add a title
plt.title('Scatter Plot of Cost vs Number of Components Affected')

# Show the plot
plt.show()
```

# Question 2 
**Q2**: You need to randomly sample 5 rows from the DataFrame, and output the results as a DataFrame.

- Write your code in the cell below
- Do not delete the comment line ### Your code here ###

```python
# Randomly sample 5 rows from the DataFrame
sampled_df = df.sample(n=5, random_state=1)

# Output the results
print(sampled_df)
```

# Question 3 
**Q3**: You need to output the minimum valule and the maximum value of the feature `n_comp_affected`, using the Pandas DataFrame min() and max() method.

- output them as an ndarray object named `X_tr`. 
- the shape of `X_tr` should be (2, 1).
- Do not delete the comment line ### Your code here ###

```python
### Q3: Your code here ###
# Find min and max of 'n_comp_affected'
min_value = sampled_df['n_comp_affected'].min()
max_value = sampled_df['n_comp_affected'].max()

# Create ndarray X_tr with shape (2, 1)
X_tr = np.array([[min_value], [max_value]])

# Output X_tr
print(X_tr)
```

Now we can train the linear regression model using the dataset.

```python
# Do not modify this cell
from sklearn.linear_model import LinearRegression

# For simplicity, we are using all the data for training in this example, and evaluation is not considered.
lin_reg = LinearRegression()
lin_reg.fit(df['n_comp_affected'].values.reshape(-1, 1), df['cost'].values)
```


# Question 4
**Q4**: Use your model to compute the target value (the 'cost' column) of `X_tr` and output the prediction. You should:

- output the prediction as a 1-d array named `y_pred_tr`
- Do not delete the comment line ### Your code here ###

```python
### Q4: Your code here ###
# Predict the cost values using the trained model
y_pred_tr = lin_reg.predict(X_tr) 

# predicts corresponding value of y, for the input values of X. which is the min and max of X


# Output the predictions
print("Predictions for X_tr:")
print(y_pred_tr)
```

# Question 5 
**Q5**: Plot the dataset and your linear regression model in the same figure using matplotlib. You should:

- show your name and student ID above the figure
- use RED DOTS to represent the training data
- use BLUE LINE to represent the model
- Do not delete the comment line ### Your code here ###

```python
### Q5: Your code here ###
# Extracting data for plotting
X_data = df['n_comp_affected'].values.reshape(-1, 1)
y_data = df['cost'].values

# Plotting the training data
plt.scatter(X_data, y_data, color='red', label='Training Data')

# Plotting the linear regression line
# Define the range of n_comp_affected values for prediction
X_range = np.linspace(X_data.min(), X_data.max(), 100).reshape(-1, 1)
y_pred = lin_reg.predict(X_range)

plt.plot(X_range, y_pred, color='blue', linewidth=2, label='Linear Regression Model')

# Adding labels and title
plt.xlabel('n_comp_affected')
plt.ylabel('cost')
plt.title('Linear Regression Model')
plt.legend()

# Display the plot
plt.show()
```
